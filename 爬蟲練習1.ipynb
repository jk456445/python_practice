```python
import requests
from bs4 import BeautifulSoup
```

```python
r= requests.get("https://www.dada-master.com/lesson?page=2")
print(r.text)
```

上面那個是把網頁的html抓下來

```python
soup= BeautifulSoup(r.text,"html.parser")  #將網頁資料以html.parser解析器來解析文檔並存入 soup 
print(soup)  #印出至畫面
```

```python
sel = soup.select("div.item-text h3.b-title") #選取 div.item-text h3.b-title ，並存入sel
print(sel)
```

```python
find_result= soup.find_all("h3") #搜尋所有符合條件的HTML節點，回傳符合條件的節點列表
print(find_result)
```
