```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from ipywidgets import interact
```

```python
def f(x):
    print(x*2)
```

```python
f(3)
```

```python
interact(f, x=3)
```

```python
interact(f, x='hi')
```

```python
interact(f, x=["taipei","taichung","tainan"])
#可以做出下拉式選單
#要做成字典也可以
```

```python
x=np.linspace(0,10,100)
```

```python
x
```

```python
plt.plot(x)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.plot(x[np.sin(x)>0], np.sin(x)[np.sin(x)>0], 'o') #特別標出sin(x)>0的地方，[]裡面的表示範圍，x和y軸都要標
```
